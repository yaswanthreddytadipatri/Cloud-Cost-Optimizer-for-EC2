# ***Cloud Cost Optimizer***

## ***Data preparation***

### Using dataset.csv

```python
import pandas as pd
df = pd.read_csv('dataset.csv')
df
```

### Selecting Required Columns

```python
dataset = df[['API Name','Memory','vCPUs','Linux On Demand cost','Linux Reserved cost','Linux Spot Minimum cost','Linux Spot Maximum cost']]
dataset
```

### Sorting Data based on Memory, vCPUs, Linux On Demand cost, Linux Reserved cost, Linux Spot Minimum cost and Linux Spot Maximum cost

```python
Sorted_data = dataset.sort_values(by=['Memory', 'vCPUs', 'Linux On Demand cost', 'Linux Reserved cost', 'Linux Spot Minimum cost', 'Linux Spot Maximum cost']).reset_index(drop=True)
Sorted_data
# Sorted_data.to_csv('sorted_data.csv', index=False)
```

## ***Taking Input Data***

```python
CPU = 10
Memory = 19
instance_type = 'i2.4xlarge'
price_Linux_On_Demand_Cost = Sorted_data.loc[Sorted_data['API Name'] == instance_type, 'Linux On Demand cost'].iloc[0]
price_Linux_Reserved_Cost = Sorted_data.loc[Sorted_data['API Name'] == instance_type, 'Linux Reserved cost'].iloc[0]
price_Linux_Spot_Minimum_Cost = Sorted_data.loc[Sorted_data['API Name'] == instance_type, 'Linux Spot Minimum cost'].iloc[0]
price_Linux_Spot_Maximum_Cost = Sorted_data.loc[Sorted_data['API Name'] == instance_type, 'Linux Spot Maximum cost'].iloc[0]
print(price_Linux_On_Demand_Cost)
print(price_Linux_Reserved_Cost)
print(price_Linux_Spot_Minimum_Cost)
print(price_Linux_Spot_Maximum_Cost)
```

### Getting the index value by taking thresold as 80%

```python
Index_Memory = Sorted_data.loc[(Sorted_data['Memory'] * 0.8 >= Memory) &(Sorted_data['vCPUs'] * 0.8 >= CPU)].index[0]
print(Sorted_data.loc[Index_Memory])
```

# ***Output***

```python
count = 1 #Counter value to display the number of suggestions

for i in range(Index_Memory, Index_Memory-5,-1):
    print("\tSuggestion: ",count)
    print(Sorted_data.loc[i],'\n')
    
    # Prining the price saved for Linux On Demand Cost
    price_Linux_On_Demand_Cost_Index_Memory = Sorted_data.loc[i, 'Linux On Demand cost']
    Price_saved_Linux_On_Demand_Cost = round(float(price_Linux_On_Demand_Cost) - float(price_Linux_On_Demand_Cost_Index_Memory), 2)
    print('For Linux on Demand price will saved by $', round(Price_saved_Linux_On_Demand_Cost, 2), '\hour | $', round(Price_saved_Linux_On_Demand_Cost*24, 2), '\day | $', round(Price_saved_Linux_On_Demand_Cost*24*30, 2), '\month | $', round(Price_saved_Linux_On_Demand_Cost*24*365, 2), '\year')

    # Prining the price saved for Linux Reserved Cost
    price_Linux_Reserved_Cost_Index_Memory = Sorted_data.loc[i, 'Linux Reserved cost']
    Price_saved_Linux_Reserved_Cost = round(float(price_Linux_Reserved_Cost) - float(price_Linux_Reserved_Cost_Index_Memory), 2)
    print('For Linux Reserved price will saved by $', round(Price_saved_Linux_Reserved_Cost, 2), '\hour | $', round(Price_saved_Linux_Reserved_Cost*24, 2), '\day | $', round(Price_saved_Linux_Reserved_Cost*24*30, 2), '\month | $', round(Price_saved_Linux_Reserved_Cost*24*365, 2), '\year')

    # Prining the price saved for Linux Spot Minimum Cost
    price_Linux_Spot_Minimum_Cost_Index_Memory = Sorted_data.loc[i, 'Linux Spot Minimum cost']
    Price_saved_Linux_Spot_Minimum_Cost = round(float(price_Linux_Spot_Minimum_Cost) - float(price_Linux_Spot_Minimum_Cost_Index_Memory), 2)
    print('For Linux Spot Minimum Price will saved by $', round(Price_saved_Linux_Spot_Minimum_Cost, 2), '\hour | $', round(Price_saved_Linux_Spot_Minimum_Cost*24, 2), '\day | $', round(Price_saved_Linux_Spot_Minimum_Cost*24*30, 2), '\month | $', round(Price_saved_Linux_Spot_Minimum_Cost*24*365, 2), '\year')

    print('\n-----------------------------------------------------------------------------------------------------\n')
    count += 1
```
